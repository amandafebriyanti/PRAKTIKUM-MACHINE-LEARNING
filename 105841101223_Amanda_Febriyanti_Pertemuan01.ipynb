```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

np.random.seed(42)

print("--- Memulai Tugas Praktikum ---\n")
```

Pada bagian ini saya mengimport library yang diperlukan. Dimana NumPy yang saya gunakan untuk mengoperasikan angka, kemudian Pandas untuk membuat tabel data, terus Matplotlib untuk membuat grafik, sedangkan Scikit-Learn untuk mengambil dataset wine. Seed diatur agar angka acak tetap sama setiap saya menjalankan program

```python
# 1. TUGAS NUMPY

print("1. Tugas NumPy")

# Membuat array berisi 10 angka acak antara 1 sampai 100
angka = np.random.randint(1, 101, 10)

# Menampilkan isi array
print("Data Angka:", angka)

# Menghitung rata-rata dari angka
print("Rata-rata:", np.mean(angka))

# Mencari nilai terbesar
print("Nilai Maksimum:", np.max(angka))

# Mencari nilai terkecil
print("Nilai Minimum:", np.min(angka))
print("\n")
```

Pada bagian ini saya membuat array yang berisi 10 angka acak antar 60 sampai 100. Kemudian selanjutnya saya menghitung nilai rata-rata, nilai tinggi, dan nilai terendah menggunakan fungsi bawaan NumPy. Hal ini menunjukan bahwa NumPy sangat membantu dalam melakukan perhitungan statistik dasar dengan cepat

```python
# 2. TUGAS PANDAS

print("2. Tugas Pandas")

# Membuat data nilai mahasiswa dalam bentuk dictionary
data_mhs = {
    'Nama': ['Hizrah', 'Hasriana', 'Manda', 'Yusuf', 'Rasma', 'Reski', 'Pahila', 'Andika', 'April', 'Indah' ],
    'Nilai': [85, 70, 95, 88, 60, 75, 92, 80, 82,78]
}

# Mengubah dictionary menjadi DataFrame
df_mhs = pd.DataFrame(data_mhs)

# Menampilkan seluruh data mahasiswa
print("Data Nilai Mahasiswa:")
print(df_mhs.to_string(index=False))

# Mengambil 3 mahasiswa dengan nilai tertinggi

top_3 = df_mhs.nlargest(3, 'Nilai')

print("\n2b. 3 Nilai Tertinggi:")
print(top_3.to_string(index=False))
print("\n")
```

Selanjutnya pada bagian ini saya membuat DataFrame yang berisi nama dan nilai mahasiswa. DataFrame memudahkan pengelolahan data dalam bentuk tabel. Saya juga menggunakan fungsi nlargest() untuk menampilkan tiga mahasiswa dengan nilai paling tinggi.

```python
# 3. TUGAS VISUALISASI
# ==================================================

print("3. Menampilkan Grafik...")


# Membuat grafik batang untuk membandingkan nilai mahasiswa
plt.subplot(1, 2, 1)
plt.bar(df_mhs['Nama'], df_mhs['Nilai'], color='skyblue')
plt.title('Grafik Batang Nilai')
plt.xticks(rotation=45)

# Membuat grafik pie untuk melihat proporsi nilai
plt.subplot(1, 2, 2)
plt.pie(df_mhs['Nilai'], labels=df_mhs['Nama'], autopct='%1.1f%%', startangle=140)
plt.title('Proporsi Nilai (Pie)')

# Mengatur tata letak agar rapi
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

Selanjutnya saya membuat dua jenis grafik. Garafik batang yang digunakan untuk membandingkan nilai setiap mahasiswa secara langsung terhadap keseluruhan nilai. Sedangkan Pie chart digunakan untuk melihat proporsi nilai masing-masing mahasiswa terhadap keseluruhan nilai.

```python
# 4. TUGAS DATASET (Scikit-Learn)

# Mengambil dataset Wine dalam bentuk DataFrame
wine = load_wine(as_frame=True)
df_wine = wine.frame

print("\n4. Insight Dataset Wine:")

# Menampilkan jumlah total sampel (baris)
print(f"- Total Sampel: {df_wine.shape[0]}")

# Menampilkan jumlah total fitur (kolom)
print(f"- Total Fitur: {df_wine.shape[1]}")

# Menghitung rata-rata kadar alkohol
print(f"- Rata-rata Alkohol: {df_wine['alcohol'].mean():.2f}")

print("\n--- Selesai ---")
```

pada bagian ini saya menggunakan dataset bawaan dari Scikit-Learn yaitu dataset Wine. Dimana dataset ini memiliki sejumlah fitur kimia dari berbagai jenis wine. Di sini saya menampilkan jumlah sample, jumlah fitur, dan rata-rata kadar alkohol untuk mendapatkan gambaran umum tentang isi dataset tersebut
